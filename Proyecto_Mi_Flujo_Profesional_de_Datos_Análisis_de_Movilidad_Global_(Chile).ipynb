```
import pandas as pd

# 1. CARGA Y FILTRO (Optimización de memoria)
df = pd.read_csv('Global_Mobility_Report.csv', low_memory=False)
df = df[df['country_region'] == 'Chile'].copy()

# 2. MANEJO DE NULOS
# Se eliminan columnas 100% vacías en Chile para reducir ruido.
df = df.dropna(axis=1, how='all')
# Se rellenan sub-regiones nulas para identificar el promedio nacional.
df['sub_region_1'] = df['sub_region_1'].fillna('Nacional')

# 3. DUPLICADOS Y TIPADO
# Se eliminan registros repetidos para asegurar unicidad por fecha/lugar.
df = df.drop_duplicates()
df['date'] = pd.to_datetime(df['date'])

# 4. OUTLIERS (Detección básica)
# Se filtran valores de movilidad extremos que superen el 200% (errores de digitación).
columnas_movilidad = [col for col in df.columns if 'percent_change' in col]
for col in columnas_movilidad:
    df = df[df[col].between(-100, 200)]

# RESULTADO FINAL
print(f"Limpieza terminada. Filas finales: {len(df)}")
df.head()
```
